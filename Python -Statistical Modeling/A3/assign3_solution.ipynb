## Read dataset

```python
df = pd.read_csv('iris.csv')
df.describe()
```

```python
sepal_length = df['sepal_length']
sepal_width  = df['sepal_width']
petal_length = df['petal_length']
petal_width  = df['petal_width']
```

```python
y = petal_width
x1 = petal_length
x2 = sepal_length
```

## Standardize the dataset

```python
x1 = (x1 - x1.mean())/x1.std()
x2 = (x2 - x2.mean())/x2.std()
y = (y - y.mean())/y.std()
```

```python
 plt.scatter(y, x1)
```

# MLR Model to estimate slopes and intercept

```python
with pm.Model() as mlr:
    α = pm.Normal('α', mu=0, sigma=10)
    β1 = pm.Normal('β1', mu=0, sigma=10)
    β2 = pm.Normal('β2', mu=0, sigma=10)
    ϵ = pm.HalfCauchy('ϵ', 5)

    μ = pm.Deterministic('μ', α + β1 * x1 + β2 * x2)
    y_pred = pm.Normal('y_pred', mu=μ, sigma=ϵ, observed=y)

    trace_mlr = pm.sample(1000, tune=1000)
```

```python
posterior_mlr = az.extract(trace_mlr)
intercept = posterior_mlr['α'].mean().item()
slope1 = posterior_mlr['β1'].mean().item()
slope2 = posterior_mlr['β2'].mean().item()
print('intercept: ', intercept)
print('slope1: ', slope1)
print('slope2: ', slope2)
```

# Confounding Variable
In order to investigate for confounding effect, let us build the model with and without x2 as we did in the class

## Model 1 (Based on Petal Length)

```python
with pm.Model() as m_x1:
    α = pm.Normal('α', mu=0, sigma=10)
    β1 = pm.Normal('β1', mu=0, sigma=10)
    ϵ = pm.HalfCauchy('ϵ', 5)

    μ = pm.Deterministic('μ', α + β1 * x1)
    y_pred = pm.Normal('y_pred', mu=μ, sigma=ϵ, observed=y)

    t_x1 = pm.sample(1000, tune=1000, chains=2)
```

## Model 2 (Based on Sepal Length)

```python
with pm.Model() as m_x2:
    α = pm.Normal('α', mu=0, sigma=10)
    β2 = pm.Normal('β2', mu=0, sigma=10)
    ϵ = pm.HalfCauchy('ϵ', 5)

    μ = pm.Deterministic('μ', α + β2 * x1)
    y_pred = pm.Normal('y_pred', mu=μ, sigma=ϵ, observed=y)

    t_x2 = pm.sample(1000, tune=1000)
```

## Model 3 (Based on Petal Length and Sepal Length)

```python
with pm.Model() as m_x1x2:
    α = pm.Normal('α', mu=0, sigma=10)
    β1 = pm.Normal('β1', mu=0, sigma=10)
    β2 = pm.Normal('β2', mu=0, sigma=10)
    ϵ = pm.HalfCauchy('ϵ', 5)

    μ = pm.Deterministic('μ', α + β1 * x1 + β2 * x2)
    y_pred = pm.Normal('y_pred', mu=μ, sigma=ϵ, observed=y)

    t_x1x2 = pm.sample(1000, tune=1000)
```

```python
az.plot_forest([t_x1x2, t_x1, t_x2],
               model_names=['m_x1x2', 'm_x1', 'm_x2'],
               var_names=['β1', 'β2'],
               combined=True, colors='cycle', figsize=(12, 6))
```

As can be seen from this plot, β2 changes significantly when we bring x1 to the model. As soon as we add x1, β2 become close to zero. So the effect of x2 in defining y was confounded because of effect of x1 which hints that x1 might be the confounder variable.

# Multicollinearity

Apart from higher R values between y and x1 and y and x2, we also see high R value of 0.87 between x1 and x2 (can also be seen from the scatter plot) which indicates multicollinearity. Let us visualize the pair plot of β1 and β2.

```python
az.plot_pair(t_x1x2, var_names=['β1', 'β2'])
```

As can be seen from pair plot, β1 and β2 coefficients are also highly correlated. This implies β1 can be expressed as a linear function of β2 and vice-versa. In other words in the presense of one parameter, the other is not giving any aided information. In fact, adding both in the model will increase the uncertanity. This can also be observed from the forest plot above where HPD is wider for both the coeficients in the the combined model as compared to the model m_x1 and m_x2.

## Dealing with multicollinearity

One solution will be to drop either x1 or x2. As R value of y with x1 (0.96) is higher than the R value of y with x2 (0.81), therefore, we should drop x2.
