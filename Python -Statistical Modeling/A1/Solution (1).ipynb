```python
import pymc3 as pm
```

# Our data from the above discription 

```python
total_trails=150
no_observed_6=39
```

# We need to define model in order to estimate the probablity <br>

We'll define model with different priors as mentioned in our A1

# let's define the model with <br>


\begin{align*}
\theta & \sim\mathrm{Uniform}\left(\ lower=0,\ upper=1\right)\\
y & \sim\mathrm{Bin}\left(n=1,p=\theta\right)
\end{align*}

```python


with pm.Model() as Model_1:
    # we need prior information
    p=pm.Uniform("p",lower=0, upper=1)                                                      # Uniform prior with lower and upper parameter
    #we need likleyhood for modeling the above data
    Likelyhood=pm.Binomial("Likelyhood", p=p,n=total_trails,observed=no_observed_6)
    idata_U=pm.sample(1000)
```

```python
idata_U['p'].mean()
```

So with    \begin{align*}
\theta & \sim\mathrm{Uniform}\left(\ lower=0,\ upper=1\right)\\
\end{align*}

## Our p estimation is 0.2627274679892616
