```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
arr5d = np.array([[5, 10, 2, 14, 9],
                  [3, 7, 8, 1, 6],
                  [13, 8, 4, 11, 0],
                  [15, 4, 8, 6, 2],
                  [10, 1, 9, 3, 5]])

arr5d2 = np.array([[5, 2, 2, 14, 9],
                  [0, 7, 12, 1, 6],
                  [13, 8, 4, 11, 0],
                  [15, 4, 8, 6, 2],
                  [10, 1, 9, 3, 5]])
zeros = np.where(arr5d == 0)
arr5d[zeros] = np.random.randint(1, 15, size=len(zeros[0]))
zeros = np.where(arr5d2 == 0)
arr5d2[zeros] = np.random.randint(1, 15, size=len(zeros[0]))

uniquevalues, counts = np.unique(arr5d, return_counts=True)
for value, count in zip(uniquevalues, counts):
  print(f"values: {value}, counts: {count}")
firstsorted = arr5d[arr5d[:, 0].argsort()]
secondsorted = arr5d2[arr5d2[:, 0].argsort()]
print(firstsorted)
array1stats = {
'Mean': np.nanmean(arr5d),
'Median': np.nanmedian(arr5d),
'Minimum': np.nanmin(arr5d),
'Maximum': np.nanmax(arr5d)
}
print(array1stats)
array2stats = {
'Mean': np.nanmean(arr5d2),
'Median': np.nanmedian(arr5d2),
'Minimum': np.nanmin(arr5d2),
'Maximum': np.nanmax(arr5d2)
}
print(array2stats)


greater = arr5d > arr5d2
greaterequal = arr5d >= arr5d2
less = arr5d < arr5d2
lessequal = arr5d <= arr5d2

print(greater)

plt.bar(uniquevalues, counts)
plt.title('value v count')
plt.xlabel('value')
plt.ylabel('count')
plt.show()
```
