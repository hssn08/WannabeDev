```python
import random

def calc_fitness(chromo):

   fitness = 0
   for i in range(5):
       x, y = chromo[i]

       fitness += 1
       if x > 0:
           fitness += 1
       if x < 5:
           fitness += 1
       if y > 0:
           fitness += 1
       if y < 5:
           fitness += 1
   return fitness

def selection(pop, fitness):

   total_fitness = sum(fitness)
   probabilities = [f/total_fitness for f in fitness]


   r = random.random()
   cumulative_prob = 0
   for i, prob in enumerate(probabilities):
       cumulative_prob += prob
       if r < cumulative_prob:
           chromo1 = pop[i]
           break

   r = random.random()
   cumulative_prob = 0
   for i, prob in enumerate(probabilities):
       cumulative_prob += prob
       if r < cumulative_prob:
           chromo2 = pop[i]
           break

   return chromo1, chromo2

def crossover(chromo1, chromo2):

   crossover_point = random.randint(0, 4)
   offspring1 = chromo1[:crossover_point] + chromo2[crossover_point:]
   offspring2 = chromo2[:crossover_point] + chromo1[crossover_point:]
   return offspring1, offspring2

def mutation(chromo):

   mut_idx = random.randint(0, 4)
   mut_chromo = chromo.copy()
   x, y = mut_chromo[mut_idx]
   y = max(0, y - 1)
   mut_chromo[mut_idx] = (x, y)
   return mut_chromo


def main():

   population = [[(random.randint(0, 5), random.randint(0, 5)) for i in range(5)] for j in range(20)]
   print("Initial Population:", population)

   for i in range(100):
       fitness = [calc_fitness(chromo) for chromo in population]

       chromo1, chromo2 = selection(population, fitness)


       offspring1, offspring2 = crossover(chromo1, chromo2)


       offspring1 = mutation(offspring1)
       offspring2 = mutation(offspring2)


       worst_idx1 = fitness.index(min(fitness))
       worst_idx2 = fitness.index(min(fitness[:worst_idx1] + fitness[worst_idx1+1:]))
       population[worst_idx1] = offspring1
       population[worst_idx2] = offspring2


       best_idx = fitness.index(max(fitness))
       best_chromo = population[best_idx]
       best_fitness = fitness[best_idx]
       print(f'Generation {i+1}: Best Chromosome = {best_chromo}, Fitness = {best_fitness}')

main()
```
